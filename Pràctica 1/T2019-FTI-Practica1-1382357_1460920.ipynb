```python
# EXERCISE 1a: Implement divisibility rule for 9
#
# Function UAB_divisible_by_9.
# * Parameter n: Integer to check (>= 0)
# * Returns: boolean, true if n is divisible by 9, false otherwise.
#

def UAB_divisible_by_9(n):

    result = True

    if(n>0):
        digitos=list(str(n))
        suma=0
        flag=False
        result=False

        while(flag==False):
            for i in digitos:
                suma+=(int(i))

            digitos=list(str(suma))

            if(len(str(suma)) == 1):
                flag=True
                if(suma==9):
                    result=True
            else:
                suma=0


    return result
```

```python
# EXERCISE 1b: Implement divisibility rule for 11
#
# Function UAB_divisible_by 11.
# * Parameter n: Integer to check (>= 0)
# * Returns: boolean, true if n is divisible by 11, false otherwise.
#

def UAB_divisible_by_11(n):
    
    result = True
    
    if(n<0):
        result=False
    
    if(n>0):
        result=False
        digitos=list(str(n))
        pares=digitos[0::2]
        impares=digitos[1::2]
        sumaP=0
        sumaI=0
        resultado=0
        flag=False
        
        while flag==False:
            for i in impares:
                sumaI+=int(i)
            for j in pares:
                sumaP+=int(j)
            
            resultado=abs(sumaP-sumaI)
            
            if(resultado==0):
                flag=True
                result=True
            else:
                resultado=list(str(resultado))
                if(len(resultado))==1:
                    flag=True
                else:
                    pares=resultado[0::2]
                    impares=resultado[1::2]
                    resultado=0
                    sumaP=0
                    sumaI=0

            
    return result
```

```python
# EXERCISE 2a: Compute gcd between two integers using Euclid's Algorithm
#
# Function UAB_gcd_with_euclides.
# * Parameter m: Integer to check (>= 0)
# * Parameter n: Integer to check (>= 0)
# * Returns: gcd between the integers m and n.
#

def UAB_gcd_with_euclides(m, n):
    
    result=max(m,n)
    
    if((m>0) and (n>0)):
        result=m%n
        while result!=0:
            m=n
            n=result
            result=m%n
        result=n
    
    return result
```

```python
# EXERCISE 3a: Compute Euler's totient function
#
# Function UAB_gcd_with_euclides.
# * Parameter n: Integer (>0)
# * Returns: phi(n), the number of number of positive integers less than n that are coprime to n
#

def UAB_euler_phi(n):
    
    result = 1
    flag=True
    j=1
    
    if(n==0):
        result=0
    if(n>1):
        lista=[]
        factor=list(n.factor())
        for i in factor:
            lista.append(pow(i[0],i[1]))
        
        if(len(lista)>1):
            while flag==True and j<=len(lista)-1:
                if(UAB_gcd_with_euclides(lista[0],lista[j])!=1):
                    flag=False
                j+=1

            if(flag):
                for i in lista:
                    result*=(UAB_euler_phi(i))
        else:
            result=(pow(factor[0][0],factor[0][1]))-(pow(factor[0][0],(factor[0][1]-1)))
    
    return result
```
